```python
#url for playstation 4 slim
url_bin ="https://www.ebay.com/sch/i.html?_nkw=playstation+4+slim+&+_BIN=1+&_ipg=100&rt=nc&LH_BIN=1&_pgn="
user_agent = {'User-agent': 'Mozilla/5.0'} 
#current_directory = os.getcwd()
```

For the first 10 pages of 100 items/page, save all the URLs of sponsored items' pages to the file "sponsored.txt"  and all the URLs of non-sponsored items' pages to the file "non-sponsored.txt" in the same directory as your code. (One URL per line in each file)   


```python
filename_1 = 'sponsored.txt'
filename_2 = 'non-sponsored.txt'
file_1 = open(filename_1,'a+')
file_2 = open(filename_2,'a+')
file_1.truncate(0)
file_2.truncate(0)
#with open(filename_1, 'a+') as file_1:
for pg in range(1,11):
        time.sleep(10)
        url_search = url_bin + str(pg)
        #print(url_search)
        user_agent = {'User-agent': 'Mozilla/5.0'} 
        ab = requests.get(url_search)
        soup = BeautifulSoup(ab.text, 'html.parser') 
        items = soup.select("#srp-river-results > ul > li.s-item")
        for item in items:
                a = item.find("a")
                if not a:
                    continue
                
                text = item.find("span").get_text()
                if re.search('.*S.*P.*O.*N.*S.*O.*R.*E.*D.*',text):  
                    url_simplify = re.sub("(.*)\\?.*", r"\1", a['href'])           
                    #print(url_simplify)
                    file_1.write(url_simplify + '\n')
                if not re.search('.*S.*P.*O.*N.*S.*O.*R.*E.*D.*',text):  
                    file_2.write(a['href'] +'\n')    
file_1.close()
file_2.close()
```

Write a separate piece of code that loops through the pages you downloaded in (c) and opens and parses them into a Python or Java xxxxsoup-object. Identify and select:

seller name, seller score, item price, # items sold, best offer available, title, returns allowed, shipping price, condition (e.g., used, new, like new, seller refurbished, ...).

```python
col_names =  ["sponsored","title","seller_name", "seller_score", "item_price", "items_sold","best_offer","shipping_cost","returns_policy","returnboolean","condition"]
ebay_db = pd.DataFrame(columns = col_names)
```

```python
ebay_db
```

```python
path = os.getcwd()
list_nonsponsored = os.listdir(path)
for x in list_nonsponsored:
    filename_ns = x
    with open(filename_ns, 'r') as file:
        text = file.read()
        soup = BeautifulSoup(text, 'html.parser')
        
        #For identifying the product price 
        try :
            item_p = soup.find('span', attrs={'id':'prcIsum'})  
            price_p = (item_p.get_text()).strip()
            cur, price_int = price_p.split('$')
            p= re.findall(r'\b\d+\b', price_p)
            price_f = int(str(p[0]) + str(p[1]))
        except Exception as e:
                pass
            

        #For identifying the product title      
        try :
            item_t = soup.find('h1', attrs={'id':'itemTitle'})
            title_p = (item_t.get_text()).strip()
        except Exception as e:
                pass
            
            
        #For printing the seller information
        try:
                sename = soup.find('span',attrs={'class' : 'mbg-nw'}).get_text().strip()
        except Exception as e:
                pass
            
        try:
                seller_score = soup.find("span",{'class' : 'mbg-l'}).find('a').get_text().strip()
                seller_score = int(seller_score)
        except Exception as e:
                pass   
            
            
        #For identifying the product condition      
        try :
            item = soup.find('div', attrs={'id':'vi-itm-cond'})
            condition_p = (item.get_text())
        except Exception as e:
                pass
        
        #For identifying the product returns policy
        try :
            returns = soup.find_all('span',attrs = {'class':'w2b-sgl'})
            for x in returns:
                z = re.search(".*[r|R]eturns.*",x.get_text())
                if(z):
                    returns_policy = x.get_text()
                    if(returns_policy == 'No returns'):
                        return_p = 0
                    else:
                        return_p = 1   
        except Exception as e:
                pass
        
        #For identifying the items sold
        try:
            items_sold = soup.find('a',attrs={'class':'vi-txt-underline'})
            if items_sold is None:
                    items_s = 0
            else:
                    items_1 = soup.find('a',attrs={'class':'vi-txt-underline'}).get_text()
                    i = re.findall(r'\b\d+\b', items_1)
                    items_s = int(str(i[0]))
        except Exception as e:
                pass
        
        #For identifying the bestsellers
        try:
            best_offer = soup.find('div',attrs = {'class':'vi-bbox- u-f1L lable boLable'})
            if best_offer is None:
                best_s = None
            else:
                    best_s =soup.find('div',attrs = {'class':'vi-bbox-dspn u-flL lable boLable'}).get_text()
        except Exception as e:
                pass
        #rint(best_s)
        
        #For calculating the shipping cost
        try:
            shipping_price = soup.find('span',attrs={'id' : 'shSummary'}).find("span").get_text()
            if(re.search(".*[f|F][r|R][e|E][e|E].*",shipping_price)):
                  ship_cost = 0
            elif(re.search(".*[v|V]aries.*",shipping_price)):
                   ship_cost = None
            else:
                   ship_cost= shipping_price
                   
        except Exception as e:
                pass
        
        
        ebay_db = ebay_db.append(pd.DataFrame(
              {'sponsored' : 'Non-sponsored','title':title_p,'seller_name':sename, 'seller_score':seller_score, 'item_price':price_f,'items_sold':items_s,
              'best_offer':best_s,'shipping_cost':ship_cost,'returns_policy':returns_policy,'returnboolean':return_p,
             'condition':condition_p}, index=[0]), ignore_index=True)        








#ay_db = ebay_db.append(pd.DataFrame(
#              {'title':title_p,'seller_name':sename, 'seller_score':seller_score, 'item_price':price_f 'items_sold':items_s,
#              'best_offer':best_s,'shipping_cost':shipping-price,'returns_policy':returns_policy,'returnboolean':return_p,
#             'condition':condition_p}, index=[0]), ignore_index=True)        
```

Use your code script to connect to SQL (either MySQL, MariaDB, or SQLite. Do NOT use SQL GUI or command terminal). Create a database and name it "eBay". Save the information of items in (d)  into a single table named "eBay_items"

```python
#connecting to mysql
connection = mysql.connector.connect(host='localhost',
                                         user='root')
mydb = connection.get_server_info()
print("Connected to MySQL Server version ", mydb)
```

```python
#create database ebay
mycursor = connection.cursor()
mycursor.execute("CREATE DATABASE if not exists ebay")
```

```python
#Use data base ebay
sql_db_select = "use ebay;"
mycursor.execute(sql_db_select)
```

```python
#create table
mycursor.execute("CREATE TABLE if not exists ebay_items (sponsored VARCHAR(50),title VARCHAR(500),seller_name VARCHAR(255), seller_score INT,item_price INT,items_sold INT, best_offer VARCHAR(5),shipping_cost VARCHAR(150),returns_policy VARCHAR(50), returnboolean INT,condi VARCHAR(50))")
```

```python
engine = create_engine("mysql+pymysql://{user}@localhost/{db}"
                       .format(user="root",
                               db="ebay"))
```

```python
#Transfer ebay_db dataframe to sql table ebay_items
ebay.to_sql(name = 'ebay_items', con=engine, if_exists='replace', index=False)
connection.commit()
```

###### Summary Stats for Item_price(values in cents)

```python
mycursor.execute("select sponsored,condition_p,round(avg(item_price),0) as Mean,round(min(item_price),0) as Minimum,round(max(item_price),0) as maximum from ebay_items group by sponsored,condition_p;")
result = mycursor.fetchall()
print(tabulate(result, headers=['sponsored', 'condition_p','Mean','Minimum','maximum'], tablefmt='psql'))    
```
